## 项目背景
中学教育是学生成长学习过程中至关重要的一环，而和教育相关的很多被我们忽视的重要信息往往埋藏在数据当中，因此对教育数据的分析可以对学生学习、教师教学和学校管理的方法起到优化提升的作用。该项目的数据来自阿里巴巴天池的“数智教育“数据可视化创新大赛(https://tianchi.aliyun.com/competition/entrance/231704/information) 。数据集一共有七份，内容包括：近五年各班各学科的教师信息、当前在校学生详细信息、学生考勤信息、学生考试成绩信息、学生消费信息。这套数据最吸引我的地方在于它有实际应用场景（中学），且分析的结果能够有针对性的输出（受众不同），和教学管理贴合得十分紧密。

## 项目价值
该项目的受众主要分为三个层面：学校（校领导）、班级（班主任/科任教师）、学生（家长/教师/学生）。

项目的价值体现在：
1. 通过建立学校看板帮助校领导了解全校学生人口信息、学生成绩以及教师情况；
2. 通过建立班级看板帮助班主任了解班级综合情况、班级的相对优势和劣势学科，以及每个学科的学生成绩分布；
3. 通过搭建学习成绩预测模型来帮助分析与学生综合成绩相关的因素。

#### 考勤种类归纳
由考勤类型数据集可以看到考勤类型看似有15种，而实际上由于考勤机的更换（根据id和年份推测），考勤中有的类型互相重复，因此可以得到进一步归纳。

```python
df_kaoqintype
```

我们可以把考勤问题分为几大类：
* 迟到：100100、100200、9900100、100300（此类型无数据）
* 校服校徽：200100、9900200
* 早退：200200、9900300
* 课间操请假： 300200 （仅2013-14学年度有数据）
* 常规考勤信息：100000、200000（无数据）、300000、300100、9900400、9900500

```python
#summarise attendance data into a few large categories
df_attendance['type'] = 0
mask_late = (df_attendance['control_task_order_id'] == 100100) | (df_attendance['control_task_order_id'] == 100200) | (df_attendance['control_task_order_id'] == 9900100)
df_attendance.loc[mask_late,'type'] = '迟到'

mask_uniform = (df_attendance['control_task_order_id'] == 200100) | (df_attendance['control_task_order_id'] == 9900200)
df_attendance.loc[mask_uniform,'type'] = '校服校徽'

mask_leave = (df_attendance['control_task_order_id'] == 200200) | (df_attendance['control_task_order_id'] == 9900300)
df_attendance.loc[mask_leave,'type'] = '早退'

mask_exercise = (df_attendance['control_task_order_id'] == 300200) 
df_attendance.loc[mask_exercise,'type'] = '课间操请假'

mask_normal = (df_attendance['control_task_order_id'] == 100000) | (df_attendance['control_task_order_id'] == 300000) | (df_attendance['control_task_order_id'] == 300100) | (df_attendance['control_task_order_id'] == 9900400) | (df_attendance['control_task_order_id'] == 9900500)
df_attendance.loc[mask_normal,'type'] = '常规'
```

从学校角度来看，最关键的指标是学生高考成绩。 但由于数据中不含高考成绩，那么科目最全、最接近高考成绩的就是“十校联考”成绩了。因浙江省在2017年对高考做了改革，不再分文理科，所以这里只参考改革以后的数据。因新高考的一段分数线是在全省前20%的成绩处切线，那么有参考意义的指标就是十校联考中学校前20%的分数线了。因此，我计算了2017和18年的20%分数线和同比趋势。

```python
#get scores from the college entrance exam simulation tests
df_score_ten = df_score[(df_score['EXAM_KIND_ID']==7)]
```

```python
#get score from the test that includes all subjects in 2017
df_score_ten_2017 = df_score_ten[(df_score_ten['exam_term']== '2016-2017-1') & (df_score_ten['exam_number']== 268)]
#get score from the test that includes all subjects in 2018
df_score_ten_2018 = df_score_ten[(df_score_ten['exam_term']== '2017-2018-2')]
```

```python
def thresh_score(df,year):    
    #get students who attended all 6 subjects exams
    subnum = df.groupby(by=['mes_StudentID'])['mes_Score'].count()
    id_sub6 = subnum[subnum==6].index.tolist()
    df_score_sub6 = df[df['mes_StudentID'].isin(id_sub6)]
    #calculate total scores from each of these students
    df_score_sum = df_score_sub6.groupby(by=['mes_StudentID'])['mes_Score'].sum()
    #sort scores from high to low
    df_score_sort = df_score_sum.sort_values(ascending=False).reset_index()
    #get the top 20% threshold
    total_num = df_score_sum.count()
    thresh_index = int(round(total_num/5 - 1))
    thresh = df_score_sort.loc[thresh_index,'mes_Score']
    print("{}年十校联考前20%分数线为{}。".format(year,thresh))
```

```python
thresh_score(df_score_ten_2017, 2017)
thresh_score(df_score_ten_2018, 2018)
```

## 班级指标和看板

对于班级来说，最重要的指标是班级各科平均成绩, 而班级成绩则应当以年级平均成绩为参考。为便于比较每个班级各学科的成绩，我以各年级各科的平均成绩为单位计算出各学科权重指数，从而每个年级都有一个固定的平均总成绩权重指数和平均学科成绩权重指数。通过年级成绩权重指数，我又推算出了每个班级各学科的权重指数。因为各班级平均总成绩指数均和年级保持一致，所以我们可以更清晰地比较各班各学科的优势和劣势。在成绩数据集中，和班级信息相挂钩的只有2018-19年第一学期期中和高三五校、十校联考的数据，高一和高二的期中考试数据齐全，而高三的期中考试只有语文和数学两科的记录，十校联考中则有除语文、数学以外其他学科的记录，因此我以这学期的期中和高三十校联考成绩为例进行计算。

```python
# remove data from redundant exam
df_stu_score_reduced = df_stu_score[df_stu_score['EXAM_KIND_NAME']!= "五校联考"]
df_stu_score_reduced.head()
```

```python
#calculate average score per subject by grade 
df_score_grade = df_stu_score_reduced.groupby(['grad','mes_sub_name'])[['mes_Score']].mean()
df_score_grade = df_score_grade.reset_index()
df_score_grade = df_score_grade.rename(columns={"mes_Score" : "mes_score_grad"})
```

```python
#calculate average score per subject by class
df_score_class = df_stu_score_reduced.groupby(['grad','cla_Name','mes_sub_name'])[['mes_Score']].mean()
df_score_class = df_score_class.reset_index()
df_score_class = df_score_class.rename(columns = {"mes_Score": "mes_score_class"})
```

```python
#join the two dfs
df_score_weighted = df_score_grade.merge(df_score_class, how="left",on=["grad","mes_sub_name"])
df_score_weighted.head()
```

```python
#find the lowest subject score in each grade
min_grad1 = df_score_weighted.groupby('grad')['mes_score_grad'].min()[0]
min_grad2 = df_score_weighted.groupby('grad')['mes_score_grad'].min()[2]
min_grad3 = df_score_weighted.groupby('grad')['mes_score_grad'].min()[1]
```

```python
#calculate weighted grade average score by subject
df_score_weighted.loc[df_score_weighted['grad'] =="高一", 'score_grad_weighted'] = df_score_weighted['mes_score_grad']/min_grad1
df_score_weighted.loc[df_score_weighted['grad'] =="高二", 'score_grad_weighted'] = df_score_weighted['mes_score_grad']/min_grad2
df_score_weighted.loc[df_score_weighted['grad'] =="高三", 'score_grad_weighted'] = df_score_weighted['mes_score_grad']/min_grad3
```

```python
#calculate the total weighted index per grade
mask = df_score_weighted.groupby(['grad','mes_sub_name'])['score_grad_weighted'].mean().reset_index()
mask = mask.groupby('grad').sum()
```

```python
#add the grade total index to the df
df_score_weighted.loc[df_score_weighted['grad'] =="高一", 'score_grad_w_total'] = mask['score_grad_weighted'][0]
df_score_weighted.loc[df_score_weighted['grad'] =="高二", 'score_grad_w_total'] = mask['score_grad_weighted'][2]
df_score_weighted.loc[df_score_weighted['grad'] =="高三", 'score_grad_w_total'] = mask['score_grad_weighted'][1]
```

```python
# calculate the total score for each class
total_score_class = df_score_weighted.groupby('cla_Name')['mes_score_class'].sum().reset_index()
total_score_class = total_score_class.rename(columns={"mes_score_class":"score_class_total"})
```

```python
df_score_weighted = df_score_weighted.merge(total_score_class, how="left", on="cla_Name")
```

```python
#calculate the weighted index for each subject per class
df_score_weighted['score_class_weighted'] = df_score_weighted['mes_score_class']/df_score_weighted['score_class_total']*df_score_weighted['score_grad_w_total']
df_score_weighted.head()
```

```python
df_score_weighted.to_csv('education_data/df_score_weighted.csv')
```

### 添加消费信息
下面我通过学生的消费数据计算了一下从2018年7月1日至2019年1月30日（数据所涵盖的时间段）间每个学生最近一次消费的时间间隔、消费频率和消费额度。

```python
df_consumption
```

首先计算距离最近一次消费的时间间隔：

```python
#create a snapshot date using one day after the most recent date in the dataset
snapshot = df_consumption['date'].max() + pd.Timedelta(days=1)
#calculate the recency by using snapshot date minus the most recent date of consumption
r = df_consumption.groupby('bf_StudentID').agg({'date': lambda x: (snapshot - x.max()).days})
r = r.rename(columns = {'date':'recency'}).reset_index()
```

然后计算消费频率：

```python
#calculate frequency of consumption
f = df_consumption.groupby('bf_StudentID').size().reset_index()
f = f.rename(columns={0:'frequency'})
```

然后计算消费额度：

```python
#calculate each student's sum of consumption
m = df_consumption.groupby('bf_StudentID').sum().MonDeal.reset_index()
m = m.rename(columns={'MonDeal':'monetary_value'})
```

```python
# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); sns.distplot(r['recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); sns.distplot(f['frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); sns.distplot(m['monetary_value'])
# Show the plot
plt.show()
```

从图中可以看到，虽然学生们最近一次的消费时间集中在数据集结束时间前不久，但消费频率和消费金额的分布都比较广泛。
为便于衡量，下面我将这三个消费维度的数据分别划分为5等分，并为每一区间赋分：5分代表最近一次消费间隔短/消费频率高/金额高，而1分则反之。

```python
r_labels = range(5,0,-1)
f_labels = range(1,6)
m_labels = range(1,6)

r_groups = pd.qcut(r['recency'], q=7, labels=r_labels, duplicates='drop')
r = r.assign(R=r_groups.values).astype(int)
f_groups = pd.qcut(f['frequency'], q=5, labels=f_labels)
f = f.assign(F=f_groups.values).astype(int)
m_groups = pd.qcut(m['monetary_value'], q=5, labels=m_labels)
m = m.assign(M=m_groups.values).astype(int)
```

```python
rfm = r.merge(f, on='bf_StudentID').merge(m, on='bf_StudentID')
```

```python
df_stu_score_atten_consu = df_stu_score_atten.merge(rfm, how='left', on='bf_StudentID')
```

### 搭建预测模型

为了尽量减少过拟合的可能性，需要对模型的预测结果做交叉验证以及确定其评估标准。这里我选择用adjusted R-squared（修正决定系数）来评估模型。该系数可以在显示自变量对因变量解释能力的同时又避免单纯由因变量的数量而导致的过拟合。为了能够对模型进行调优，这里还需要把已经分好的训练数据集再划分为训练集和验证集(validation set)。

```python
#function for parameter tuning
def tuning(model, X, y, parameters, score_func=None, n_folds=5, n_jobs=-1):
    #shuffle and split data to train and validation set
    kf = KFold(n_folds, shuffle=True, random_state=0).get_n_splits(X.values)
    #optimize parameters
    if score_func:
        gs = GridSearchCV(model, param_grid=parameters, cv=kf, n_jobs=n_jobs, scoring=score_func)
    else:
        gs = GridSearchCV(model, param_grid=parameters, cv=kf, n_jobs=n_jobs)
    gs.fit(X.values, y.values)
    print("BEST: {}\n {}\n {}".format(gs.best_params_, gs.best_score_, gs.best_estimator_))
```

```python
#function for model fitting
def fitting(model, X, y, parameters=None, score_func=None, n_folds=5, n_jobs=-1):
    #shuffle and split data to train and validation set
    kf = KFold(n_folds, shuffle=True, random_state=0).get_n_splits(X.values)
    #calculate validation set score
    rmse= np.sqrt(-cross_val_score(model, X.values, y.values, scoring=score_func, cv = kf)) 
    model.fit(X.values,y.values)
#     R2 = cross_val_score(model, X.values, y.values, scoring="r2", cv = kf)
#     n = round(df_X_train.shape[0]/5)
#     p = 13
#     adj_R2 = 1-(1-R2)*(n-1)/(n-p-1)
    return(rmse)
```

```python
#models to be used
linear_model = LinearRegression()
lasso_model = Lasso() 
```

```python
#lasso parameters
lasso_params = {
    'alpha': [0.0001],
    'random_state' : [0]
}
```

```python
#function to check feature importance

def feature_impor(model, X):
    importance_list = model.coef_
    name_list = list(X.columns.values[1:])
    importance_list, name_list = zip(*sorted(zip(importance_list, name_list)))
    plt.barh(range(len(name_list)),importance_list,align='center')
    plt.yticks(range(len(name_list)),name_list)
    plt.xlabel('Coef')
    plt.ylabel('Features')
    plt.show()
```

#### 模型一（无单科成绩）：

```python
#fit linear regression model with training set
score = fitting(linear_model, df_X_train1, y_train, score_func='neg_mean_squared_error')
print("\nscore: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
feature_impor(linear_model,df_X_train1)
```

```python
#check linear regression model prediction
y_predict = linear_model.predict(df_X_test1.values)
score = np.sqrt(mean_squared_error(y_test.values, y_predict))
round(score, 4)
```

```python
#fit lasso model training set
score = fitting(lasso_model, df_X_train1, y_train, lasso_params, 'neg_mean_squared_error')
print("\nscore: {:.4f} ({:.4f})\n".format(score.mean(), score.std()))
feature_impor(lasso_model, df_X_train1)
```

```python
#check lasso model prediction
y_predict = lasso_model.predict(df_X_test1.values)
score = np.sqrt(mean_squared_error(y_test.values, y_predict))
round(score, 4)
```
